```python
def loadmodels():
    try:
        #give path to the gender detection model object - keras object
        gender_modelpath = r"Models\vggmodelthird.h5"

        #give path to the face recognition pickle object
        facerecog_picklepath = "Pickles\encodings_latest.pickle"

        #reading the models
        print("loading the gender detection model")
        gender_model = load_model(gender_modelpath)

        print("loading the face recognition model")
        facerecog_model = pickle.loads(open(facerecog_picklepath, "rb").read())
        
        return(facerecog_model,gender_model)
        
    except Exception as ex:
        print(ex)
```

```python
def putNameOverFace(current_frame, top, bottom, left, right, name):
    
    # draw the predicted face name on the image
    cv2.rectangle(current_frame, (left, top-20), (right, bottom),(0, 255, 0), 2)

    y = top - 20 if top - 20 > 20 else top + 20

    #Add name over face rectange
    cv2.putText(current_frame, name, (left, y), cv2.FONT_HERSHEY_SIMPLEX,0.75, (0, 255, 0), 2)
    
    
    #crop the face and return it
    face_cropped = np.copy(current_frame[top-50:bottom+20, left-20:right+20])
    
    return(face_cropped)
    
```

```python
def detectGenderFromPositions(currentrgb_frame, top, bottom, left, right,show_conf):
    global togglevalue
    
    # crop the detected face region
    face_crop = np.copy(currentrgb_frame[top-20:bottom, left:right])

    # preprocessing for gender detection model
    face_crop = cv2.resize(face_crop, (96,96))
    face_crop = face_crop.astype("float") / 255.0
    face_crop = img_to_array(face_crop)
    face_crop = np.expand_dims(face_crop, axis=0)

    # apply gender detection on face
    conf = gender_model.predict(face_crop)[0]

    # get label with max accuracy
    idx = np.argmax(conf)
    label = classes[idx]
    
    if togglevalue == 1:
        label_per = "{}".format(label)
    else:
        label_per = "{}: {:.2f}%".format(label, conf[idx] * 100)

    #print(label)
    Y = bottom - 10
    startX = left

    # write label and confidence above face rectangle
    cv2.putText(currentrgb_frame, label_per, (startX, Y),  cv2.FONT_HERSHEY_SIMPLEX,0.6, (255, 255, 0), 1)
    
    return(label)
```

```python
def image_capture(queue):
    
    try:
        global teamnames_excel_lst, excel_data,teamname_flag
        
        team_name = list(set(person_info["TeamName"]))
        
        #file path
        outfolder_path = os.getcwd()+"\\TeamOutput\\"
        
        #create directory if not exists
        if not os.path.exists(outfolder_path): 
            os.makedirs(outfolder_path)
            
        team_folder_path = os.getcwd() + "\\TeamOutput\\" + team_name[0]    
        
        #check if folder exists with same team name
        if not (os.path.exists(team_folder_path) == True or os.path.exists(team_folder_path.lower()) == True):
            os.makedirs(team_folder_path)
            
            #save the photos in the folder.
            for image,name in zip(lst_croppedfaces, person_info["PersonName"]):
                im = PilImage.fromarray(image)
                im.save(team_folder_path +"\\"+name+".jpg")
                
        #appending the data to excel
        ExcelWrapper.appendDataToExcel(person_info)
        
        #log message
        messagelbl.config(text="Team details saved sucessfully",fg='green')
        
        excel_data = ExcelWrapper.readDataFromExcel()
        teamnames_excel_lst = list(set((excel_data)["TeamName"]))
        
        teamname_flag=0
        Teamname_entry.delete(0,END)
        Teamname_entry.update()
                
    except Exception as ex:
        print(ex)
        print("error")
```

```python
def messagedisplay(msg):
    if msg == 'Message box':
        messagelbl.config(text=msg,fg='gray75')
    else:
        messagelbl.config(text=msg,fg = 'red',wraplength = 700)
```

```python
# Camera change  function
def cameraChange():
    global cap
    if entry_ip.get()!="":
        if cam_value==1:
            cap.release()
            cap = cv2.VideoCapture(0)
            cap.set(cv2.CAP_PROP_FRAME_WIDTH, ws/2);
            cap.set(cv2.CAP_PROP_FRAME_HEIGHT, hs/2);
            messagedisplay('Message box')
        else:
            try:
                cap.release()
                cap=cv2.VideoCapture("https://"+str(entry_ip.get())+"/video")
                cap.set(cv2.CAP_PROP_FRAME_WIDTH, ws/2);
                cap.set(cv2.CAP_PROP_FRAME_HEIGHT, hs/2);    
                time.sleep(2)
                if not cap.isOpened():   
                    cap = cv2.VideoCapture(0)#checks for the opening of camera
                    cap.set(cv2.CAP_PROP_FRAME_WIDTH, ws/2);
                    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, hs/2);
                    messagedisplay('Cannot Connect to IP camera!!')
            except Exception as ex:
                messagedisplay(ex)
    else:
        messagedisplay('Please Enter IP address')
        
# function for web came and phone cam change
def sel():    
    global cam_value
    cam_value =   int(cam_var.get())
   
    if cam_value==2:
        entry_ip.configure({"background": "snow"})
        entry_ip['state']='normal'
        IP_btn['state']='normal'
        
    if cam_value==1:
        entry_ip.configure({"background": "gray74"})
        entry_ip['state']='disabled'
        IP_btn['state']='disabled'
        cameraChange()

def close_window():
    root.destroy()
    
# funtion for video capture button enable/ disable
def btnstatus(dmcount,dfcount,name):
    
    messagedisplay("Message box")
    
    global excel_data

    if(int(tkvar1.get())==dmcount and int(tkvar2.get())==dfcount and teamname_flag==1):
        if name == "Unknown":
            messagedisplay("Unknown persons present in the frame. Adjust the camera position or Ensure only audited persons will be present in the frame")
        else:
            if name in list(excel_data.PersonName):
                messagedisplay(name + " is already present in an another team.")
            else:
                messagedisplay('Message box')
                greenBtn['state'] = 'normal'
    else:
        greenBtn['state']='disabled'

# funtion foe showing the percentage of gender detection
def toggle():
    global togglevalue
    togglevalue=1
    if togglebtn.config('text')[-1] == 'Show':
        togglebtn.config(text='Hide',bg="tomato")
        togglevalue=0
    else:
        togglebtn.config(text='Show',bg="pale green")
        togglevalue=1

# dynamic display of count of men and women in real time       
def dynamic_display(dmcount,dfcount):
        
    panel = tk.Label(rightFrame, text = "Male Count            ",bg='gray1',font=('Times', '12','bold'),
                     justify=LEFT,fg="RosyBrown1")
    panel.grid(row=3,column=0,padx=5,pady=15)
    panel = tk.Label(rightFrame, text = str(dmcount),bg='gray1',font=('Times', '14','bold'),
                     justify=LEFT,fg="snow")
    panel.grid(row=3,column=1,padx=5,pady=15)
    panel = tk.Label(rightFrame, text = " Female Count          ",bg='gray1',font=('Times', '12','bold'),
                     justify=LEFT,fg="RosyBrown1")
    panel.grid(row=4,column=0,padx=5,pady=15)
    panel = tk.Label(rightFrame, text = str(dfcount),bg='gray1',font=('Times', '14','bold'),
                     justify=LEFT,fg="snow")
    panel.grid(row=4,column=1,padx=5,pady=15)


        
global teamname_flag
teamname_flag=0 # 1 for success for team name validation

def teamname_validation():
    global teamname_flag,teamname_entered,teamnames_excel_lst
    
    #global messagelbl
    teamname_entered=Teamname_entry.get()
    teamnames_excel_lst = [x.lower() for x in teamnames_excel_lst]
    
    if teamname_entered.isalnum() is not True:
        messagedisplay('Please enter only Alpha Numeric for Team Name')
        teamname_flag=0
    else:
        if teamname_entered.lower() in teamnames_excel_lst:
            messagedisplay('Team Name ' + teamname_entered + " already taken. Enter a different name")
            teamname_flag=0
        else:
            messagedisplay('Message box')
            teamname_flag=1

        
# NIRANJAN CODE INTEGRATION
def show_gender_vid():
    
    try:
        if not cap.isOpened(): #checks for the opening of camera
            print("Can't open the camera")

        # read frame from webcam 
        flag, frame = cap.read()
        frame = cv2.flip(frame, 1)
       # time.sleep(0.5)

        if flag is None:
            print("Major error!")
            exit()

        elif flag:
            global last_frame,person_info,lst_croppedfaces,excel_data
            last_frame = frame.copy()
        
        dm_count=0
        df_count=0
        gender_lst,person_name_lst,lst_croppedfaces,team_name_lst = [],[],[],[]
        person_info = {}

        # convert the input frame from BGR to RGB 
        rgb_frame = cv2.cvtColor(last_frame, cv2.COLOR_BGR2RGB)
        

        # Find all the faces and face encodings in the current frame of video
        face_locations = face_recognition.face_locations(rgb_frame,number_of_times_to_upsample=1)
        face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

        #loop over all the encodings
        for encoding,location in zip(face_encodings,face_locations):

            # attempt to match each face in the input image to our known encodings
            matches = face_recognition.compare_faces(facerecog_model["encodings"],encoding, tolerance=0.4)
            name = "Unknown"

            top, right, bottom, left = location[0], location[1], location[2], location[3]

            # check to see if we have found a match
            if True in matches:
                # find the indexes of all matched faces then initialize a dictionary to count the total number of times each face was matched
                matchedIdxs = [i for (i, b) in enumerate(matches) if b]
                counts = {}
                # loop over the matched indexes and maintain a count for each recognized face face
                for i in matchedIdxs:
                    name = facerecog_model["names"][i]
                    counts[name] = counts.get(name, 0) + 1

                # determine the recognized face with the largest number of votes
                name = max(counts, key=counts.get)


            #Finding the gender
            gender = detectGenderFromPositions(rgb_frame, top, bottom,left,right,togglevalue)
            
            if gender == "man":
                dm_count += 1
            elif gender == "woman":
                df_count += 1
                
            #put names to faces
            cpface = putNameOverFace(rgb_frame, top, bottom,left,right, name) 
            
            #checking the btn status    
            btnstatus(dm_count,df_count,name)
            
            dynamic_display(dm_count,df_count)
            
            if name != "Unknown":
                lst_croppedfaces.append(cpface)
                person_name_lst.append(name)
                gender_lst.append(gender)
                if teamname_flag == 1:
                    team_name_lst.append(teamname_entered)
                
        person_info['PersonName'] = person_name_lst
        person_info['Gender'] = gender_lst
        person_info['TeamName'] = team_name_lst
        

        img = PilImage.fromarray(rgb_frame)
        imgtk = ImageTk.PhotoImage(image=img,)
        lmain.imgtk = imgtk
        lmain.configure(image=imgtk)
        lmain._image_cache = imgtk  # avoid garbage collection
        lmain.after(30,show_gender_vid) 

    except Exception as ex:
        print(ex)
```
